```python
import csv

# Nombres de los archivos
input_file = '/content/drive/MyDrive/entrenamientos/DATASETS/latidos_normal.csv'
output_file = 'filtered_signals_normal.csv'

# Leer y filtrar las señales
with open(input_file, mode='r', newline='') as infile, open(output_file, mode='w', newline='') as outfile:
    reader = csv.reader(infile)
    writer = csv.writer(outfile)

    # Leer el encabezado y escribirlo en el archivo de salida
    header = next(reader)
    writer.writerow(header)

    for row in reader:
        # Asumimos que cada fila representa una señal y que la longitud de la señal es la cantidad de datos en la fila
        signal_data = row[1:]  # Saltamos el primer elemento si es un identificador u otro tipo de metadato
        if len(signal_data) > 128:
            writer.writerow(row)
```
