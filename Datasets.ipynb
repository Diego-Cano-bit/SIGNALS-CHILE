```python
import os
import csv
import shutil
import pandas as pd
from sklearn.model_selection import train_test_split

# Rutas a las carpetas
input_folder = '/content/drive/MyDrive/entrenamientos/DATASETS/Señales_traslapadas/Chf'
train_folder = '/content/drive/MyDrive/entrenamientos/DATASETS/Div_traslapadas/TRAIN_70'
test_folder = '/content/drive/MyDrive/entrenamientos/DATASETS/Div_traslapadas/TEST_20'
val_folder = '/content/drive/MyDrive/entrenamientos/DATASETS/Div_traslapadas/VAL_10'

# Crear las carpetas de salida si no existen
os.makedirs(train_folder, exist_ok=True)
os.makedirs(test_folder, exist_ok=True)
os.makedirs(val_folder, exist_ok=True)

# Obtener la lista de archivos CSV
csv_files = [f for f in os.listdir(input_folder) if f.endswith('.csv')]

# Dividir los archivos en entrenamiento, prueba y validación
train_files, temp_files = train_test_split(csv_files, test_size=0.3, random_state=42)
val_files, test_files = train_test_split(temp_files, test_size=2/3, random_state=42)

# Función para concatenar archivos CSV
def concatenate_and_save(files, src_folder, dest_file):
    df_list = []
    for file in files:
        file_path = os.path.join(src_folder, file)
        df = pd.read_csv(file_path)
        df_list.append(df)
    concatenated_df = pd.concat(df_list, ignore_index=True)
    concatenated_df.to_csv(dest_file, index=False)

# Concatenar y guardar los archivos en las carpetas correspondientes
concatenate_and_save(train_files, input_folder, os.path.join(train_folder, 'train.csv'))
concatenate_and_save(test_files, input_folder, os.path.join(test_folder, 'test.csv'))
concatenate_and_save(val_files, input_folder, os.path.join(val_folder, 'val.csv'))

print(f'Total de archivos: {len(csv_files)}')
print(f'Archivos de entrenamiento: {len(train_files)}')
print(f'Archivos de prueba: {len(test_files)}')
print(f'Archivos de validación: {len(val_files)}')
```
