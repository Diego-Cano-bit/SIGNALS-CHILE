```python
import csv
import matplotlib.pyplot as plt

def graficar_fila_csv(archivo_csv, num_fila):
    # Lee el archivo CSV
    with open(archivo_csv, newline='') as csvfile:
        lector_csv = csv.reader(csvfile)
        filas = list(lector_csv)

        # Verifica que la fila solicitada exista
        if num_fila < 0 or num_fila >= len(filas):
            print(f"La fila {num_fila} está fuera de los límites del archivo CSV.")
            return

        # Extrae los datos de la fila
        fila = filas[num_fila]

        # Asume que la primera fila contiene los encabezados
        encabezados = filas[0]
        datos = [float(valor) for valor in fila]

        # Grafica la fila
        plt.figure(figsize=(10, 5))
        plt.plot(encabezados, datos, )
        plt.title(f'Gráfico de la fila {num_fila} del archivo {archivo_csv}')
        plt.xlabel('Encabezados')
        plt.ylabel('Valores')
        plt.grid(True)
        plt.show()

# Ejemplo de uso
archivo_csv = '/content/drive/MyDrive/normal_signals.csv'
num_fila = 1  # Cambia esto al número de fila que se requiere graficar (empezando desde 0)
graficar_fila_csv(archivo_csv, num_fila)
```
